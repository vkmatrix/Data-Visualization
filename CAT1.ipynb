# SEABORN COLOR PALETTES

```python
import numpy as np
import pandas as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#  QUALITATIVE COLOR PALETTE
cpal=sns.color_palette()
sns.palplot(cpal)
plt.show()
```

```python
# SEQUENTIAL COLOR PALETTE

cp=sns.color_palette(palette="Blues")
sns.palplot(cp)
plt.show()
```

```python
#  DIVERGING COLOR PALETTE

pal=sns.color_palette(palette="BrBG",n_colors=7,desat=0.8)
sns.palplot(pal)
plt.show()
```

# SEABORN FIGURE AESTHETICS

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#  without seaborn figure aesthetics
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    plt.plot(x,np.sin(x))
        
sinplot()
```

```python
#  with seaborn figure aesthetics
sns.set(style="whitegrid")
sinplot()
plt.show()
```

# Removing axes spines

```python
df=sns.load_dataset('iris')
df
```

```python
sns.set(style='ticks')
sns.scatterplot(df.sepal_length,df.sepal_width,data=df)
plt.title('this plot is with top and right spines');
```

```python
sns.set(style='ticks')
sns.scatterplot(df.sepal_length,df.sepal_width,data=df)
sns.despine()
plt.title('this plot is with removed top and right spines');
```
