# Customizing plots using seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def sinplot():
    x=np.linspace(0,10,100)
    y=np.sin(x)
    plt.plot(x,y)
    plt.title('sine plot')
    plt.xlabel('x')
    plt.ylabel('sin(x)');
sinplot()
```

```python
sns.set_style('whitegrid')
sinplot()
```

```python
sns.set_style('darkgrid')
sinplot()
sns.despine()
```

#  Violin Plot

```python
df=sns.load_dataset('iris')
df
```

```python
# One variable only
sns.set_style('ticks')
sns.violinplot(y=df['sepal_length'])
sns.despine();
```

```python
# One variable and severa classes of other variable
sns.set_style('whitegrid')
sns.violinplot(x=df['species'],y=df['sepal_length'])
sns.despine();
```

```python
# horizontal
sns.violinplot(y=df['species'],x=df['sepal_length']);
```

```python
# several variables 
sns.violinplot(data=df.iloc[:,0:2],linewidth=4);
```

```python
sns.set_style('dark')
sns.boxplot(y=df.iloc[:,0]);
```

```python
sns.boxplot(x=df['species'],y=df['sepal_length']);
```

```python
sns.boxplot(y=df['species'],x=df['sepal_length']);
```

```python
sns.boxplot(data=df.iloc[:,:2]);
```

# Multiplots 

```python
data=sns.load_dataset('tips')
g=sns.FacetGrid(data,col='sex',hue='smoker',height=4,aspect=1.5)
g.map(sns.scatterplot,"total_bill","tip")
g.add_legend();
```

```python
g1=sns.FacetGrid(data=data,col="sex",height=4,aspect=1.5)
g1.map(sns.histplot,"total_bill");
```

```python
g2=sns.FacetGrid(data=data,col="smoker",hue="sex",height=4,aspect=1.5)
g2.map(sns.scatterplot,"total_bill","tip");
g2.add_legend();
```

#  Plotting Bivariate Relationships and Visualizing Pairwise Rel

```python
df=pd.read_csv('AgeIncome.csv')
df
```

```python
#  KDE PLOT
sns.kdeplot(x=df['age'],y=df['income'],shade=True);
```

```python
#  HEXBIN PLOT
sns.set_style('ticks')
sns.jointplot(data=df,x='age',y='income',kind='hex',color='green');
plt.show()
```

```python
# SCATTERPLOT
sns.jointplot(data=df,x='age',y='income',kind='scatter',color='orange');
```

```python
# Visualizing Pairwise relationships
iris=sns.load_dataset('iris')
sns.pairplot(data=iris,kind="scatter",palette="husl",diag_kind="kde",hue="species")
```

```python
sns.pairplot(data=df,kind="scatter",diag_kind='kde',palette="husl");
```

# Regression Plot

```python
sns.lmplot(x='total_bill',y='tip',data=data,hue='sex',markers=['o','+'])
```

# UNIT 3 BOKEH

# 5 charts

```python
from bokeh.plotting import figure,output_notebook,show
```

```python
# line chart
output_notebook()
fx=figure()
fx.line(x=[1,2,3],y=[1,2,3],line_width=5)
show(fx)
```

```python
# Scatter plot
output_notebook()
fig=figure()
fig.scatter([1,2,3],[1,2,3],color=['red','blue','green'],size=10)
show(fig)
```

```python
# bar chart
output_notebook()
fg=figure()
fg.vbar(x=[1,4,7],width=1.5,bottom=0,top=[1.2,3.4,5.6],color=['red','green','blue'])
show(fg)
```

```python
# ELLIPSE CHART  (BUBBLE CHART)
from math import pi
output_notebook()
p=figure(height=600,width=600)
p.ellipse(x=[1,2,3],y=[1,2,3],width=[0.2,0.5,0.1],height=0.3,angle=pi/2,color=['violet','red','green'])
show(p)
```

```python
# WEDGE AND ARC GRAPH
output_notebook()
p1=figure()
p1.arc(x=[1,2,3],y=[1,2,3],start_angle=0.4,end_angle=4.8,color='blue',radius=0.2)
show(p1)
```
