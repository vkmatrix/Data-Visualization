```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df
```

# LINE CHART

```python
averages = df.groupby('species').mean()
averages.T.plot(marker='o')
plt.xlabel('Features')
plt.ylabel('Average Value')
plt.legend()
plt.show()
```

# SCATTER PLOT

```python
species=df['species'].unique()
for s in species:
    data=df[df['species']==s]
    plt.scatter(data['petal_length'],data['petal_width'],label=s)
plt.title('PETAL LENGTH VS PETAL WIDTH')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
plt.show()
```

# BAR CHART

```python
species=df['species'].unique()
avg_sepal_width=df.groupby("species")["sepal_width"].mean()
plt.bar(species,avg_sepal_width,color=['red','green','blue'])
plt.title("Bar Chart: Sepal Width by Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Width")
plt.show()
```
