# Plotting Bivariate Relationships

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('AgeIncome.csv')
df
```

# KDE PLOT

```python
sns.kdeplot(x='age', y='income',data=df,shade=True)
plt.title('KDE AGE VS INCOME')
plt.show();
```

# SCATTER PLOT 

```python
sns.jointplot(x='age', y='income',data=df,kind='scatter')
plt.show()
```

# HEXBIN PLOT

```python
sns.jointplot(x='age',y='income',data=df,kind='hex')
plt.show()
```

# Visualizing Pairwise Relationships

```python
iris=sns.load_dataset('iris')
sns.pairplot(data=iris,kind='scatter',hue='species',diag_kind='kde',palette='husl')
```

```python
sns.pairplot(df,kind='scatter',diag_kind='kde',palette='husl');
```
